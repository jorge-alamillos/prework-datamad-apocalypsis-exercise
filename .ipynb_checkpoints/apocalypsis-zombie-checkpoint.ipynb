# Apocalipsis zombie

![](./images/thewalkingdead.jpeg)

Cargas un arsenal de armas y una mochila con cargadores de munición de tamaño variable para algunos tipos de armas. ¿Cuántos zombies puedes matar antes de tener que echar a correr por tu vida? Deberás imprimir el resultado de cada caso que corresponde al número de balas que tienes de las armas que cargas.  

ALGORITMO: Por cada arma diferente que cargas, tienes que comprobar si portas cargadores para ella. Si esto es afirmativo deberás sumar el total de balas que podrás disparar. 

## Caso 0: Ejemplo simple

**Output esperado** -> 30

```python
armas = ['pistola','escopeta']

cargadores = {
    'pistola': [10, 10], 
    'escopeta': [2, 2, 2, 2, 2]
}

# your code
balas_disparadas = 0
balas_en_cargador = 0
   
for arma in cargadores:
    
    while len(cargadores[arma]) > 0 and balas_en_cargador == 0:
        balas_en_cargador = cargadores[arma][0]
        del cargadores[arma][0]
        print("---------------------------")
        print("Balas en el cargador:", balas_en_cargador)
        while balas_en_cargador != 0:
            balas_en_cargador += - 1
            balas_disparadas += 1
            print("Disparo la", arma,". Quedan",balas_en_cargador, "balas en el cargador")
            if balas_en_cargador == 0 and len(cargadores[arma]) > 0:
                print("Cargador vacío, recargo la", arma)
            elif balas_en_cargador == 0 and len(cargadores[arma]) == 0:
                print("No quedan cargadores de", arma,". Cambio de arma.")
           
print("---------------------------")  
print("Toca correr,no tengo más armas. He eliminado", balas_disparadas, "zombies.")
```
